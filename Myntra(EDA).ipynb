ðŸ“Œ **Step 5: Univariate Analysis**

```python
df['OriginalPrice (in Rs)'].hist(bins=50, figsize=(8, 5))
plt.title('Price Distribution')
plt.show()
```

- Prices are right-skewed (most products priced low/medium, few are premium).

```python
sns.boxplot(x=df['OriginalPrice (in Rs)'])
plt.title("Boxplot of Prices")
plt.show()
```

- Boxplot shows outliers in high price range.

```python
plt.figure(figsize=(10,5))
ax = sns.barplot(x = df['BrandName'].value_counts().head(10).index, y=df['BrandName'].value_counts().head(10).values, hue=None, legend=False, palette= 'rocket')
plt.xlabel('Brand Name')
plt.ylabel('Frequency')
plt.title('Top 10 Popular Brands')
ax.set_xticklabels(ax.get_xticklabels(), rotation = 90)
plt.tight_layout()
plt.show()
```

- The Top 10 popular brands are Pothys, Roadster, Kalini and so on.

```python
sns.countplot(x='category_by_Gender', data=df)
plt.title("Distribution by Gender")
plt.show()
```

- More products for Womenâ€™s fashion compared to Menâ€™s.

ðŸ“Œ **Step 5: Bivariate Analysis**

```python
sns.scatterplot(x='OriginalPrice (in Rs)', y='Ratings', data =df, alpha=0.5)
plt.title("Price vs Rating")
plt.show()
```

-  Higher prices donâ€™t always guarantee higher ratings.

```python
top_categories = df['Individual_category'].value_counts().head(10).index
plt.figure(figsize=(12,6))
sns.boxplot(x='Individual_category', y='OriginalPrice (in Rs)', data=df[df['Individual_category'].isin(top_categories)])
plt.xticks(rotation=45)
plt.title("Top 10 Categories vs Price")
plt.show()
```

* Categories such as Sarees, Kurtas, and Dresses exhibit higher median prices and a wider range, reflecting both affordable and premium product segments.
* In contrast, categories like Jeans, Shirts, and Tops maintain relatively lower and more consistent pricing.
* Outliers in categories like Sarees and Dresses highlight the presence of luxury fashion products, whereas Bras emerge as the most affordable and uniform category.

ðŸ“Œ**Step 6: Correlation Analysis**

```python
# Select only the numeric columns for correlation analysis, excluding 'Product_id'
numeric_df = df.select_dtypes(include=np.number).drop(columns=['Product_id'])

plt.figure(figsize=(8,6))
sns.heatmap(numeric_df.corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

* Price is positively correlated with discounted_price.

* Ratings show weak correlation with prices.

```python
sns.pairplot(df[['DiscountPrice (in Rs)', 'Ratings', 'DiscountOffer']].dropna())
plt.show()
```

* Discounts are common across all price ranges, but higher discounts appear in mid-priced items.
